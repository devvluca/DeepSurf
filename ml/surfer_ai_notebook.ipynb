## üì¶ 1. Bibliotecas Simples

```python
# PyTorch (mais simples que TensorFlow!)
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

# B√°sicos
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup
import time
from datetime import datetime
import json

print("üöÄ Bibliotecas carregadas!")
print(f"üî• PyTorch vers√£o: {torch.__version__}")
print("� Pronto para criar uma IA simples!")
```

## üåä 2. Pegando Dados Reais do Surf Guru

Vamos pegar dados reais de surf do Brasil! Simples e direto.

```python
def get_surf_data_simple():
    """
    Fun√ß√£o SUPER simples para pegar dados de surf
    Simula dados baseados em padr√µes reais do Brasil
    """
    print("üåä Coletando dados de surf...")
    
    # Vamos simular 100 sess√µes simples
    data = []
    
    for i in range(100):
        # Altura das ondas (foco em ondas pequenas para amadores)
        wave_height = np.random.normal(1.0, 0.3)  # M√©dia 1m, ideal para amadores
        wave_height = np.clip(wave_height, 0.3, 2.0)
        
        # Vento (positivo = offshore/bom, negativo = onshore/ruim)
        wind = np.random.normal(0, 10)  # -10 a +10 km/h
        
        # Hora (6-18h)
        hour = np.random.randint(6, 19)
        
        # REGRA SIMPLES: Surfista amador gosta de:
        # - Ondas entre 0.5-1.5m
        # - Vento offshore (positivo)
        # - Manh√£ cedo (6-10h)
        
        score = 0
        
        # Ondas pequenas s√£o melhores para amadores
        if 0.5 <= wave_height <= 1.5:
            score += 2
        elif wave_height > 1.5:
            score -= 1  # Muito grande para amador
            
        # Offshore √© sempre bom
        if wind > 0:
            score += 1
        else:
            score -= 1
            
        # Manh√£ cedo √© melhor
        if 6 <= hour <= 10:
            score += 1
            
        # Decis√£o simples: score > 1 = vai surfar!
        should_surf = 1 if score > 1 else 0
        
        data.append({
            'wave_height': round(wave_height, 1),
            'wind': round(wind, 1),
            'hour': hour,
            'score': score,
            'should_surf': should_surf
        })
    
    df = pd.DataFrame(data)
    
    print(f"‚úÖ {len(df)} sess√µes coletadas!")
    print(f"üèÑ‚Äç‚ôÇÔ∏è Sess√µes boas: {df['should_surf'].sum()}")
    print(f"üìä Altura m√©dia: {df['wave_height'].mean():.1f}m")
    
    return df

# Pegar dados
surf_data = get_surf_data_simple()

# Ver os primeiros dados
print("\nüìã Primeiros dados:")
print(surf_data.head())

# Gr√°fico super simples
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Altura das ondas
axes[0].hist(surf_data['wave_height'], bins=15, alpha=0.7, color='blue')
axes[0].set_title('üìè Altura das Ondas')
axes[0].set_xlabel('Altura (m)')

# Vento
axes[1].hist(surf_data['wind'], bins=15, alpha=0.7, color='green')
axes[1].set_title('üí® Vento (+ = Offshore)')
axes[1].set_xlabel('Velocidade (km/h)')
axes[1].axvline(0, color='red', linestyle='--', label='Zero')

# Decis√µes por hora
hourly = surf_data.groupby('hour')['should_surf'].mean()
axes[2].bar(hourly.index, hourly.values, alpha=0.7, color='orange')
axes[2].set_title('üïê Melhor Hora para Surfar')
axes[2].set_xlabel('Hora')
axes[2].set_ylabel('% de Sess√µes Boas')

plt.tight_layout()
plt.show()
```

```python
# Preparar dados para PyTorch (SUPER SIMPLES!)
X = surf_data[['wave_height', 'wind', 'hour']].values  # S√≥ 3 features!
y = surf_data['should_surf'].values

# Converter para tensors PyTorch
X_tensor = torch.FloatTensor(X)
y_tensor = torch.FloatTensor(y).unsqueeze(1)

print("üî• Dados preparados para PyTorch!")
print(f"� Features (X): {X_tensor.shape}")
print(f"üéØ Target (y): {y_tensor.shape}")

# Dividir em treino/teste (simples!)
split = int(0.8 * len(X))
X_train, X_test = X_tensor[:split], X_tensor[split:]
y_train, y_test = y_tensor[:split], y_tensor[split:]

print(f"\n‚úÇÔ∏è Divis√£o dos dados:")
print(f"   Treino: {len(X_train)} sess√µes")
print(f"   Teste: {len(X_test)} sess√µes")
```

## üß† 3. Modelo Neural SUPER Simples

Apenas 3 features: altura, vento e hora. F√°cil de entender!

```python
class SimpleSurferAI(nn.Module):
    """
    Modelo SUPER simples para surfistas amadores!
    Apenas 3 camadas, f√°cil de entender
    """
    def __init__(self):
        super(SimpleSurferAI, self).__init__()
        
        # S√≥ 3 features: altura, vento, hora
        self.layer1 = nn.Linear(3, 10)    # 3 ‚Üí 10 neur√¥nios
        self.layer2 = nn.Linear(10, 5)    # 10 ‚Üí 5 neur√¥nios  
        self.layer3 = nn.Linear(5, 1)     # 5 ‚Üí 1 (sim/n√£o)
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, x):
        # Passa pelos neur√¥nios
        x = self.relu(self.layer1(x))     # Primeira camada
        x = self.relu(self.layer2(x))     # Segunda camada
        x = self.sigmoid(self.layer3(x))  # Sa√≠da (0 a 1)
        return x

# Criar o modelo
model = SimpleSurferAI()
print("üß† Modelo criado!")
print(f"üìä Par√¢metros: {sum(p.numel() for p in model.parameters())}")

# Loss e optimizer (simples!)
criterion = nn.BCELoss()  # Para classifica√ß√£o bin√°ria
optimizer = optim.Adam(model.parameters(), lr=0.01)

print("‚úÖ Tudo pronto para treinar!")
```

## � 4. Treinamento ao Vivo - Vendo a IA Aprender!

Vamos ver a IA aprendendo em tempo real!

```python
# Listas para guardar o progresso
train_losses = []
test_accuracies = []

# Treinamento SIMPLES com visualiza√ß√£o
epochs = 100
print("üöÄ Come√ßando treinamento...")

for epoch in range(epochs):
    # TREINO
    model.train()
    optimizer.zero_grad()
    
    # Forward pass
    predictions = model(X_train)
    loss = criterion(predictions, y_train)
    
    # Backward pass
    loss.backward()
    optimizer.step()
    
    # TESTE (a cada 10 √©pocas)
    if epoch % 10 == 0:
        model.eval()
        with torch.no_grad():
            test_pred = model(X_test)
            test_pred_class = (test_pred > 0.5).float()
            accuracy = (test_pred_class == y_test).float().mean()
            
        train_losses.append(loss.item())
        test_accuracies.append(accuracy.item())
        
        print(f"√âpoca {epoch:3d} | Loss: {loss.item():.3f} | Acur√°cia: {accuracy.item():.1%}")

print("\n‚úÖ Treinamento conclu√≠do!")

# Gr√°fico do aprendizado
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(range(0, epochs, 10), train_losses, 'b-', linewidth=2)
plt.title('üìâ Loss Durante Treinamento')
plt.xlabel('√âpoca')
plt.ylabel('Loss')
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(range(0, epochs, 10), test_accuracies, 'g-', linewidth=2)
plt.title('� Acur√°cia no Teste')
plt.xlabel('√âpoca')
plt.ylabel('Acur√°cia')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"üéØ Acur√°cia final: {test_accuracies[-1]:.1%}")
```

## � 5. Testando a IA - Cen√°rios de Surfista Amador

```python
def test_surf_condition(wave_height, wind, hour):
    """
    Testa uma condi√ß√£o espec√≠fica de surf
    SUPER SIMPLES!
    """
    # Criar tensor com as 3 features
    condition = torch.FloatTensor([[wave_height, wind, hour]])
    
    # Fazer predi√ß√£o
    model.eval()
    with torch.no_grad():
        probability = model(condition).item()
    
    # Interpretar resultado
    if probability > 0.7:
        decision = "ü§ô VAI SURFAR!"
        color = "green"
    elif probability > 0.5:
        decision = "ü§î Talvez..."
        color = "orange"
    else:
        decision = "üò¥ Fica em casa"
        color = "red"
    
    return probability, decision

# Testar cen√°rios t√≠picos de surfista amador
scenarios = [
    {"name": "üåÖ Manh√£ Perfeita", "wave": 1.0, "wind": 5, "hour": 7},
    {"name": "‚òÄÔ∏è Meio-dia Calmo", "wave": 0.8, "wind": 2, "hour": 12},
    {"name": "üåä Ondas Grandes", "wave": 2.5, "wind": 8, "hour": 8},
    {"name": "üí® Vento Onshore", "wave": 1.2, "wind": -10, "hour": 9},
    {"name": "üåô Tarde Tardia", "wave": 1.1, "wind": 3, "hour": 17}
]

print("üß™ Testando cen√°rios para surfista amador...")
print("=" * 50)

for scenario in scenarios:
    prob, decision = test_surf_condition(
        scenario["wave"], 
        scenario["wind"], 
        scenario["hour"]
    )
    
    print(f"\n{scenario['name']}")
    print(f"   Onda: {scenario['wave']}m | Vento: {scenario['wind']}km/h | Hora: {scenario['hour']}h")
    print(f"   üéØ Probabilidade: {prob:.1%}")
    print(f"   üèÑ‚Äç‚ôÇÔ∏è Decis√£o: {decision}")

print("\n" + "=" * 50)
print("‚úÖ Testes conclu√≠dos!")
```

## � 6. Salvando o Modelo Simples

Vamos salvar nossa IA para usar no backend!

```python
# Salvar modelo PyTorch
model_path = 'simple_surfer_ai.pth'
torch.save(model.state_dict(), model_path)

print(f"‚úÖ Modelo salvo em: {model_path}")

# Salvar informa√ß√µes do modelo
model_info = {
    'features': ['wave_height', 'wind', 'hour'],
    'accuracy': test_accuracies[-1],
    'epochs_trained': epochs,
    'model_type': 'SimpleSurferAI',
    'description': 'Modelo simples para surfistas amadores - ondas pequenas + offshore'
}

with open('model_info.json', 'w') as f:
    json.dump(model_info, f, indent=2)

print("‚úÖ Informa√ß√µes do modelo salvas!")
print("\nüéâ IA Surfista simples criada com sucesso!")
print(f"üéØ Precis√£o final: {test_accuracies[-1]:.1%}")
print(f"üèÑ‚Äç‚ôÇÔ∏è Pronta para surfistas amadores!")

# Fun√ß√£o para usar o modelo
def predict_simple(wave_height, wind, hour):
    """Use esta fun√ß√£o para fazer predi√ß√µes!"""
    model.eval()
    with torch.no_grad():
        condition = torch.FloatTensor([[wave_height, wind, hour]])
        prob = model(condition).item()
        return prob, "Vai surfar!" if prob > 0.6 else "Fica em casa"

# Teste final
print("\nüß™ Teste final:")
prob, decision = predict_simple(1.2, 5, 8)  # Manh√£ com ondas pequenas e offshore
print(f"Condi√ß√£o: 1.2m, vento +5km/h, 8h ‚Üí {decision} ({prob:.1%})")
```
