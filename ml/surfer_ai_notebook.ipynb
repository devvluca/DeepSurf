## ğŸ“¦ 1. Bibliotecas Simples

```python
# PyTorch (mais simples que TensorFlow!)
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

# BÃ¡sicos
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup
import time
from datetime import datetime
import json

print("ğŸš€ Bibliotecas carregadas!")
print(f"ğŸ”¥ PyTorch versÃ£o: {torch.__version__}")
print("� Pronto para criar uma IA simples!")
```

## ğŸŒŠ 2. Pegando Dados Reais do Surf Guru

Vamos pegar dados reais de surf do Brasil! Simples e direto.

```python
def get_surf_data_simple():
    """
    FunÃ§Ã£o SUPER simples para pegar dados de surf
    Simula dados baseados em padrÃµes reais do Brasil
    """
    print("ğŸŒŠ Coletando dados de surf...")
    
    # Vamos simular 100 sessÃµes simples
    data = []
    
    for i in range(100):
        # Altura das ondas (foco em ondas pequenas para amadores)
        wave_height = np.random.normal(1.0, 0.3)  # MÃ©dia 1m, ideal para amadores
        wave_height = np.clip(wave_height, 0.3, 2.0)
        
        # Vento (positivo = offshore/bom, negativo = onshore/ruim)
        wind = np.random.normal(0, 10)  # -10 a +10 km/h
        
        # Hora (6-18h)
        hour = np.random.randint(6, 19)
        
        # REGRA SIMPLES: Surfista amador gosta de:
        # - Ondas entre 0.5-1.5m
        # - Vento offshore (positivo)
        # - ManhÃ£ cedo (6-10h)
        
        score = 0
        
        # Ondas pequenas sÃ£o melhores para amadores
        if 0.5 <= wave_height <= 1.5:
            score += 2
        elif wave_height > 1.5:
            score -= 1  # Muito grande para amador
            
        # Offshore Ã© sempre bom
        if wind > 0:
            score += 1
        else:
            score -= 1
            
        # ManhÃ£ cedo Ã© melhor
        if 6 <= hour <= 10:
            score += 1
            
        # DecisÃ£o simples: score > 1 = vai surfar!
        should_surf = 1 if score > 1 else 0
        
        data.append({
            'wave_height': round(wave_height, 1),
            'wind': round(wind, 1),
            'hour': hour,
            'score': score,
            'should_surf': should_surf
        })
    
    df = pd.DataFrame(data)
    
    print(f"âœ… {len(df)} sessÃµes coletadas!")
    print(f"ğŸ„â€â™‚ï¸ SessÃµes boas: {df['should_surf'].sum()}")
    print(f"ğŸ“Š Altura mÃ©dia: {df['wave_height'].mean():.1f}m")
    
    return df

# Pegar dados
surf_data = get_surf_data_simple()

# Ver os primeiros dados
print("\nğŸ“‹ Primeiros dados:")
print(surf_data.head())

# GrÃ¡fico super simples
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Altura das ondas
axes[0].hist(surf_data['wave_height'], bins=15, alpha=0.7, color='blue')
axes[0].set_title('ğŸ“ Altura das Ondas')
axes[0].set_xlabel('Altura (m)')

# Vento
axes[1].hist(surf_data['wind'], bins=15, alpha=0.7, color='green')
axes[1].set_title('ğŸ’¨ Vento (+ = Offshore)')
axes[1].set_xlabel('Velocidade (km/h)')
axes[1].axvline(0, color='red', linestyle='--', label='Zero')

# DecisÃµes por hora
hourly = surf_data.groupby('hour')['should_surf'].mean()
axes[2].bar(hourly.index, hourly.values, alpha=0.7, color='orange')
axes[2].set_title('ğŸ• Melhor Hora para Surfar')
axes[2].set_xlabel('Hora')
axes[2].set_ylabel('% de SessÃµes Boas')

plt.tight_layout()
plt.show()
```

```python
# Preparar dados para PyTorch (SUPER SIMPLES!)
X = surf_data[['wave_height', 'wind', 'hour']].values  # SÃ³ 3 features!
y = surf_data['should_surf'].values

# Converter para tensors PyTorch
X_tensor = torch.FloatTensor(X)
y_tensor = torch.FloatTensor(y).unsqueeze(1)

print("ğŸ”¥ Dados preparados para PyTorch!")
print(f"� Features (X): {X_tensor.shape}")
print(f"ğŸ¯ Target (y): {y_tensor.shape}")

# Dividir em treino/teste (simples!)
split = int(0.8 * len(X))
X_train, X_test = X_tensor[:split], X_tensor[split:]
y_train, y_test = y_tensor[:split], y_tensor[split:]

print(f"\nâœ‚ï¸ DivisÃ£o dos dados:")
print(f"   Treino: {len(X_train)} sessÃµes")
print(f"   Teste: {len(X_test)} sessÃµes")
```

## ğŸ§  3. Modelo Neural SUPER Simples

Apenas 3 features: altura, vento e hora. FÃ¡cil de entender!

```python
class SimpleSurferAI(nn.Module):
    """
    Modelo SUPER simples para surfistas amadores!
    Apenas 3 camadas, fÃ¡cil de entender
    """
    def __init__(self):
        super(SimpleSurferAI, self).__init__()
        
        # SÃ³ 3 features: altura, vento, hora
        self.layer1 = nn.Linear(3, 10)    # 3 â†’ 10 neurÃ´nios
        self.layer2 = nn.Linear(10, 5)    # 10 â†’ 5 neurÃ´nios  
        self.layer3 = nn.Linear(5, 1)     # 5 â†’ 1 (sim/nÃ£o)
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, x):
        # Passa pelos neurÃ´nios
        x = self.relu(self.layer1(x))     # Primeira camada
        x = self.relu(self.layer2(x))     # Segunda camada
        x = self.sigmoid(self.layer3(x))  # SaÃ­da (0 a 1)
        return x

# Criar o modelo
model = SimpleSurferAI()
print("ğŸ§  Modelo criado!")
print(f"ğŸ“Š ParÃ¢metros: {sum(p.numel() for p in model.parameters())}")

# Loss e optimizer (simples!)
criterion = nn.BCELoss()  # Para classificaÃ§Ã£o binÃ¡ria
optimizer = optim.Adam(model.parameters(), lr=0.01)

print("âœ… Tudo pronto para treinar!")
```

## � 4. Treinamento ao Vivo - Vendo a IA Aprender!

Vamos ver a IA aprendendo em tempo real!

```python
# Listas para guardar o progresso
train_losses = []
test_accuracies = []

# Treinamento SIMPLES com visualizaÃ§Ã£o
epochs = 100
print("ğŸš€ ComeÃ§ando treinamento...")

for epoch in range(epochs):
    # TREINO
    model.train()
    optimizer.zero_grad()
    
    # Forward pass
    predictions = model(X_train)
    loss = criterion(predictions, y_train)
    
    # Backward pass
    loss.backward()
    optimizer.step()
    
    # TESTE (a cada 10 Ã©pocas)
    if epoch % 10 == 0:
        model.eval()
        with torch.no_grad():
            test_pred = model(X_test)
            test_pred_class = (test_pred > 0.5).float()
            accuracy = (test_pred_class == y_test).float().mean()
            
        train_losses.append(loss.item())
        test_accuracies.append(accuracy.item())
        
        print(f"Ã‰poca {epoch:3d} | Loss: {loss.item():.3f} | AcurÃ¡cia: {accuracy.item():.1%}")

print("\nâœ… Treinamento concluÃ­do!")

# GrÃ¡fico do aprendizado
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(range(0, epochs, 10), train_losses, 'b-', linewidth=2)
plt.title('ğŸ“‰ Loss Durante Treinamento')
plt.xlabel('Ã‰poca')
plt.ylabel('Loss')
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(range(0, epochs, 10), test_accuracies, 'g-', linewidth=2)
plt.title('� AcurÃ¡cia no Teste')
plt.xlabel('Ã‰poca')
plt.ylabel('AcurÃ¡cia')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"ğŸ¯ AcurÃ¡cia final: {test_accuracies[-1]:.1%}")
```

## � 5. Testando a IA - CenÃ¡rios de Surfista Amador

```python
def test_surf_condition(wave_height, wind, hour):
    """
    Testa uma condiÃ§Ã£o especÃ­fica de surf
    SUPER SIMPLES!
    """
    # Criar tensor com as 3 features
    condition = torch.FloatTensor([[wave_height, wind, hour]])
    
    # Fazer prediÃ§Ã£o
    model.eval()
    with torch.no_grad():
        probability = model(condition).item()
    
    # Interpretar resultado
    if probability > 0.7:
        decision = "ğŸ¤™ VAI SURFAR!"
        color = "green"
    elif probability > 0.5:
        decision = "ğŸ¤” Talvez..."
        color = "orange"
    else:
        decision = "ğŸ˜´ Fica em casa"
        color = "red"
    
    return probability, decision

# Testar cenÃ¡rios tÃ­picos de surfista amador
scenarios = [
    {"name": "ğŸŒ… ManhÃ£ Perfeita", "wave": 1.0, "wind": 5, "hour": 7},
    {"name": "â˜€ï¸ Meio-dia Calmo", "wave": 0.8, "wind": 2, "hour": 12},
    {"name": "ğŸŒŠ Ondas Grandes", "wave": 2.5, "wind": 8, "hour": 8},
    {"name": "ğŸ’¨ Vento Onshore", "wave": 1.2, "wind": -10, "hour": 9},
    {"name": "ğŸŒ™ Tarde Tardia", "wave": 1.1, "wind": 3, "hour": 17}
]

print("ğŸ§ª Testando cenÃ¡rios para surfista amador...")
print("=" * 50)

for scenario in scenarios:
    prob, decision = test_surf_condition(
        scenario["wave"], 
        scenario["wind"], 
        scenario["hour"]
    )
    
    print(f"\n{scenario['name']}")
    print(f"   Onda: {scenario['wave']}m | Vento: {scenario['wind']}km/h | Hora: {scenario['hour']}h")
    print(f"   ğŸ¯ Probabilidade: {prob:.1%}")
    print(f"   ğŸ„â€â™‚ï¸ DecisÃ£o: {decision}")

print("\n" + "=" * 50)
print("âœ… Testes concluÃ­dos!")
```

## � 6. Salvando o Modelo Simples

Vamos salvar nossa IA para usar no backend!

```python
# Salvar modelo PyTorch
model_path = 'simple_surfer_ai.pth'
torch.save(model.state_dict(), model_path)

print(f"âœ… Modelo salvo em: {model_path}")

# Salvar informaÃ§Ãµes do modelo
model_info = {
    'features': ['wave_height', 'wind', 'hour'],
    'accuracy': test_accuracies[-1],
    'epochs_trained': epochs,
    'model_type': 'SimpleSurferAI',
    'description': 'Modelo simples para surfistas amadores - ondas pequenas + offshore'
}

with open('model_info.json', 'w') as f:
    json.dump(model_info, f, indent=2)

print("âœ… InformaÃ§Ãµes do modelo salvas!")
print("\nğŸ‰ IA Surfista simples criada com sucesso!")
print(f"ğŸ¯ PrecisÃ£o final: {test_accuracies[-1]:.1%}")
print(f"ğŸ„â€â™‚ï¸ Pronta para surfistas amadores!")

# FunÃ§Ã£o para usar o modelo
def predict_simple(wave_height, wind, hour):
    """Use esta funÃ§Ã£o para fazer prediÃ§Ãµes!"""
    model.eval()
    with torch.no_grad():
        condition = torch.FloatTensor([[wave_height, wind, hour]])
        prob = model(condition).item()
        return prob, "Vai surfar!" if prob > 0.6 else "Fica em casa"

# Teste final
print("\nğŸ§ª Teste final:")
prob, decision = predict_simple(1.2, 5, 8)  # ManhÃ£ com ondas pequenas e offshore
print(f"CondiÃ§Ã£o: 1.2m, vento +5km/h, 8h â†’ {decision} ({prob:.1%})")
```
